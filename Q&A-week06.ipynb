## 2. Examine the data by using visualizations.

```python
plt.figure(figsize=(10,20))

plt.subplot(4,1,1)
plt.scatter(data['International trade: Imports (million US$)'],data['Agricultural production index '],s=30,alpha=0.5, c='red')
plt.title('Plot of Agricultural production index VS International trade: Imports',fontsize=16)
plt.xlabel('International trade: Imports (million US$)')
plt.ylabel('Agricultural production index')

plt.subplot(4,1,2)
plt.scatter(data['GDP per capita (current US$)'],data['Population in thousands (2017)'],s=30, c='Blue',alpha=0.5)
plt.title('Plot of Population VS GDP per capita ',fontsize=16)
plt.xlabel('GDP per capita (current US$)')
plt.ylabel('Population in thousands (2017)')

plt.subplot(4,1,3)
plt.scatter(data['Surface area (km2)'],data['Population in thousands (2017)'],s=30, c='Green', alpha=0.5)
plt.title('Plot of Surface area VS Population',fontsize=16)
plt.xlabel('Surface area (km2)')
plt.ylabel('Population in thousands (2017)')

plt.subplot(4,1,4)
plt.scatter(data['GDP per capita (current US$)'],data['Agricultural production index '],s=30, c='violet',alpha=0.5)
plt.title('Plot of GDP per capita (current US$) VS Agricultural production index ',fontsize=16)
plt.xlabel('GDP per capita (current US$)')
plt.ylabel('Agricultural production index ')

plt.tight_layout()
```

```python
dt=data.sort_values(by='Agricultural production index ',ascending=False)

bdt=dt[['country','Region','Agricultural production index ']]
bdt1=data['Agricultural production index '].mean()
bdt2= bdt[bdt['Agricultural production index ']>=bdt1]
bdt2
```

```python
plt.figure(figsize=(14,7))
sns.countplot(x='Region',data=bdt2)
plt.title('Count of COuntries in a Region above the average Agricultural Production Index')
plt.xticks(rotation=90)
plt.show()
```

## 4. On average which region has the highest

## 5. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
std=data.groupby('Region')['GDP per capita (current US$)'].std()
std
```

```python
print('The Region that has more consistent with respect to GDP per capita (current US$) : ', std.nsmallest(1))
```
